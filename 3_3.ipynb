```python
from multiprocessing import Process
import os
import time

# 定义锁文件路径
LOCK_FILE = "test_lock"

def acquire_lock():
    """获取锁"""
    while os.path.exists(LOCK_FILE):
        print("Another server is running test.py. Waiting for 5 seconds...")
        time.sleep(5)
    # 创建锁文件
    try:
        open(LOCK_FILE, "w").close()
    except Exception as e:
        print("Failed to acquire lock:", str(e))
        # 在这里处理获取锁失败的情况

def release_lock():
    """释放锁"""
    try:
        if os.path.exists(LOCK_FILE):
            os.remove(LOCK_FILE)
    except Exception as e:
        print("Failed to release lock:", str(e))
        # 在这里处理释放锁失败的情况

def collection(option):
    """执行 collection 函数，并在执行前后处理锁"""
    try:
        acquire_lock()
        # 在这里执行 collection 函数的内容
        print(f"Process {option} is running collection...")
        time.sleep(10)  # 模拟 collection 函数的执行时间
    except Exception as e:
        print("Error in collection:", str(e))
        # 在这里处理 collection 函数执行过程中的异常
    finally:
        release_lock()

if __name__ == "__main__":
  while True: # 要有這個才會循環
    processes = []

    for i in range(1, 4):
        p = Process(target=collection, args=(i,))
        processes.append(p)
        p.start()

    for p in processes:
        p.join()
    # 每隔5分钟重新执行一次
    time.sleep(5 * 60)
```
