```python
import pandas as pd
import numpy as np

def analyze_group(group):
    data = group['col5'].reset_index(drop=True)
    col6 = group['col6'].reset_index(drop=True)
    col7 = group['col7'].reset_index(drop=True)

    results = []
    col6_records = []
    col7_records = []

    # 计算滑动窗口
    for i in range(len(data) - 2):  # 窗口大小为3
        window = data.iloc[i:i + 3]
        count_ones = (window == 1).sum()

        if (i == 0 and count_ones >= 2) or (count_ones >= 2 and window.iloc[-1] == 1):
            results.append(True)
            # 收集满足条件的col6, col7值
            indices = window[window == 1].index.tolist()
            col6_records.append(col6.iloc[indices].tolist()[::-1])
            col7_records.append(col7.iloc[indices].tolist()[::-1])
        else:
            results.append(False)

    proportion = sum(results) / len(results) if results else 0

    # 准备数据以创建结果DataFrame
    result = pd.DataFrame({
        'col1': group['col1'].iloc[0],
        'col2': group['col2'].iloc[0],
        'col3': group['col3'].iloc[0],
        'col4': group['col4'].iloc[0],
        'proportion': [proportion]
    })

    # 最大列数
    max_cols = max((len(rec) for rec in col6_records + col7_records), default=0)

    # 添加col6和col7的详细记录
    for i in range(max_cols):
        result[f'col6_{i+1}'] = [rec[i] if i < len(rec) else np.nan for rec in col6_records]
        result[f'col7_{i+1}'] = [rec[i] if i < len(rec) else np.nan for rec in col7_records]

    return result
```
